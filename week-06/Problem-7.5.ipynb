```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def confidence_int(l, N, thresh):#riemann sum from center out
    t, s = thresh * sum([(i)**l * (1-i)**(N-l) for i in np.linspace(0,1,1000)]), 0 #true normalize and set threshold
    for i, j in zip(np.linspace(l/N,1,500), np.linspace(l/N,0,500)):
            s += ((i)**l * (1-i)**(N-l)) + ((j)**l * (1-j)**(N-l)) #steps from both sides because binomial is asymmetric
            if s >= t:
                a, b = (2 * (l/N)) - i, i
                print(a, b)
                return a, b #set bounds
```

```python
l, N = 301, 600 #l = # of successes out of N trials
a, b = confidence_int(l, N, 0.1)
xi, xin = np.linspace(0,1,1001), l/N #create iterable(xi) and normalization constant(xin)
plt.plot(xi, (xi/xin)**l * ((1-xi)/(1-xin))**(N-l))
plt.axvline(x=a); plt.axvline(x=b); plt.xlim([0.4,0.6]); plt.show() #configure plot
```
