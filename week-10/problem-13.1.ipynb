## Phase Portrait & Streamlines

Draw the phase portrait in the $\theta-\Delta \bar{m}$ plane.  Add 3â€“5 streamlines.  Be sure to include some for small amplitude and large amplitude initial conditions, and discuss how small and large amplitude behavior is different.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def f1(var, t):
    return [var[1] / np.cos(var[0]), -var[0] / np.sqrt(1 + (var[0] ** 2))]

theta, dm = np.meshgrid(np.linspace(-np.pi/2, np.pi/2, 20), np.linspace(-1, 1, 20))

dmprime = f1([theta, dm], 0)

for i in [0.1,0.5,0.9]:
    path = odeint(f1, [0.0,i], np.linspace(0, 50, 200))
    plt.plot(path[:,0], path[:,1], 'b-')

plt.quiver(theta, dm, dmprime[0], dmprime[1], color = 'r')
plt.xlim([-np.pi, np.pi])
plt.ylim([-2, 2])
plt.show()
```

As the amplitude of the system increases, it spends more time at the extremeties of the system possibilities, deforming the ellipse into a more "rectangular" shape as seen above.
