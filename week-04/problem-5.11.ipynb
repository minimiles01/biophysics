```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 0, 0.3
s = np.random.normal(mu, sigma, 50000)
b = 50

x = np.linspace(min(s),max(s),num=b)
y = np.exp(-((x-mu)**2 / (2 * (sigma**2)))) \
    / (sigma * np.sqrt(2 * np.pi))
plt.hist(s,bins=b, density=True)
plt.plot(x, y, color = 'r')
plt.show()
```

```python
plt.hist(s,bins=b, density=True)
plt.plot(x, y, color = 'r')
plt.yscale('log')
plt.show()
```
